    5. How much risk is there in investing in those stocks?

    Method 1: standard deviation as a measure of risk

```python
techAdjClosing_df.head()
```

```python
area = np.pi*50 #make the scatter points bigger

sns.scatterplot(x=techAdjClosing_df.mean(),y=techAdjClosing_df.std(),s=area)
plt.xlabel('Average stock price', size=15)
plt.ylabel('STD', size=15)

#show text labels the on the scatter points
for label, x,y in zip(techAdjClosing_df.columns, techAdjClosing_df.mean(), techAdjClosing_df.std()):
    plt.annotate(
    label,
    xy = (x,y), xytext=(-10,10),
    textcoords = 'offset points', ha='left', va='bottom')
    
##risk analysis has been done with standard deviation against the daily returns.
##the lower the risk and the higher the expected returns, would be the ideal stock.
##
```

8. What is the risk level like in buying crypto currencies?

```python
crypto_df.head()
```

```python
##get the daily returns through pct_change method
crypto_df2 = crypto_df.pct_change()
crypto_df2.head()
```

```python
## drop the null values
crypto_df2 = crypto_df2.dropna()
crypto_df2.head()
```

```python
#plot the risk by expected returns plot

area = np.pi*50 #make the scatter points bigger

sns.scatterplot(x=crypto_df2.mean(),y=crypto_df2.std(),s=area)
plt.xlabel('Expected Returns', size=15)
plt.ylabel('Risk', size=15)

#show text labels the on the scatter points
for label, x,y in zip(crypto_df2.columns, crypto_df2.mean(), crypto_df2.std()):
    plt.annotate(
    label,
    xy = (x,y), xytext=(-10,10),
    textcoords = 'offset points', ha='left', va='bottom')
    
##risk analysis has been done with standard deviation against the daily returns.
##the lower the risk and the higher the expected returns, would be the ideal stock.
##
```

```python
#risk by bootstrap method
crypto_df2.head()
```
